Import the csv and add arbitrary name to the dataframe

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data.csv")

num_columns = data.shape[1]
column_names = [f'col{i+1}' if i != 1 else "type" for i in range(num_columns)]
data.columns = column_names

data.describe()
```
